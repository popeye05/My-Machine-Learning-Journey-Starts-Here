## NOW BUILDING THE REGR MODEL

```python
x = cars.drop(columns = 'Price')
y = cars['Price']
```

```python
from sklearn.model_selection import train_test_split
#This is something we done, Validation set and train set while learning the bias variace tradeoff:
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories= ohe.categories_),['name','company','fuel_type']),remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

## The DSet is very small and thus on train test split we'll get different scores, Thus well fix it
So for that we'll run a loop and take the argmax and then we'll see


```python
scores = []
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state= i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state= np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```
