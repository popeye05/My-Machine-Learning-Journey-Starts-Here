```python
#this is actually conv the description anf it returns an empty list if the value is missing
games['popular_tags'] = games['popular_tags'].apply(lambda x: x.split() if isinstance(x, str) else [])
games['all_reviews'] = games['all_reviews'].apply(lambda x: x.split() if isinstance(x, str) else [])
games['game_details'] = games['game_details'].apply(lambda x: x.split() if isinstance(x, str) else [])
games['genre'] = games['genre'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
games.head()
```

```python
games['genre']= games['genre'].apply(lambda x:[i.replace(" ","") for i in x])
games['popular_tags']= games['popular_tags'].apply(lambda x:[i.replace(" ","") for i in x])
games['game_details']= games['game_details'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
games.head(10)
```

```python
games['desc_snippet'] = games['desc_snippet'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python

games['tags'] = games['popular_tags'] + games['genre'] + games['game_details'] + games['desc_snippet']
```

```python
new_df1 = games[['name','tags']] 
```

```python
new_df1.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df1['tags']= new_df1['tags'].apply(stem)
```

```python
 new_df1['tags'] = new_df1['tags'].apply(lambda x:" ".join(x))
```

```python
new_df = new_df1.iloc[:10000]
new_df = new_df.reset_index(drop=True)
```

```python
new_df.head()
```

```python
#We spent a lot of time in Data Preprocessing, Now I'll use Vectorization and we'll use Bag of Words Algorithm.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags'])
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from  sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# sorted(list(enumerate(similarity[0])),reverse= True, key = lambda x: x[1])[1:6] This is used in game list, whereas it is computing for the 1st one
```

```python
#Now We finished the Vectorization
#Now its time for the Main function: 

def recommend(game):
    game_index = new_df[new_df['name'] == game].index[0]
    distances = similarity[game_index]
    games_list = sorted(list(enumerate(distances)),reverse= True, key = lambda x: x[1])[1:6]
    for i in games_list:
        print(new_df.iloc[i[0]]['name'])
```

```python
recommend('Grand Theft Auto IV: Complete Edition')
```
